```python
import re, json
import pprint as pp
```

## Task 1. Reading the .html file.
    Open the “spotify.html” file sent to you by email as a text file, and import all it’s content in a string. 
    

```python
with open("spotify.html", "r", encoding='utf-8') as f:
    text= f.read()      
pp.pprint(text)    
```

## Task 2. Find all the url addresses in the html.
    Extract the full list of urls for that string and save them in a json document named as ‘urls.json’.


```python
def Find_url(string):
    regex_url = 'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
    urls = re.findall(regex_url, string)
    json_string = json.dumps(urls)
    return json_string
json_url = Find_url(text)
pp.pprint(json_url)
```

```python
# Write JSON file
with open('urls.json', 'w') as json_file:
    json.dump(json_url, json_file)
```

## Task 3. Find all color codes.
    Extract the full list hex codes that represent the color of elements in the html code, 
    and save them as a list in a json document named as ‘hex.json’.


```python
def Find_hex(string):
    regex_hex = r"#[0-9a-f]{6}"
    hex_ = re.findall(regex_hex, string)
    json_string = json.dumps(hex_)
    return json_string
hex_json = Find_hex(text)
pp.pprint(hex_json)
```

```python
# Write JSON file
with open('hex.json', 'w') as json_file:
    json.dump(hex_json, json_file)
```

## Task 4. Find all artists in the html document.
	Extract the full list of artists in the html file, and save it in a list in a json document - ‘artist.json’.


```python
def Find_artist(string):
    regex_artist = "by ([\w ]+[^<]+)"
    hex_ = re.findall(regex_artist, string)
    json_string = json.dumps(hex_)
    return json_string
artist_json = Find_artist(text)
pp.pprint(artist_json)
```

```python
# Write JSON file
with open('artist.json', 'w') as json_file:
    json.dump(artist_json, json_file)
```
